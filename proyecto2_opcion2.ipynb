```python
import wfdb
import numpy as np
import matplotlib.pylab as plt
import padasip as pa
```

```python
def filter_with_NLMS(in_sig, desired_sig, n = 1, m = 0.01):
    # creation of data
    N = len(in_sig)
    filt_size =  n
    x = pa.input_from_history(in_sig,filt_size)
    d = desired_sig.reshape(len(desired_sig), 1)
    d = d[:N-filt_size+1, 0]

    f = pa.filters.FilterLMS(n=filt_size, mu=m, w="zeros")
    y, e, w = f.run(d, x)
    return y, e, w  

def filter_with_RLS(in_sig, desired_sig, n = 1, m = 0.99):
    # creation of data
    N = len(in_sig)
    filt_size =  n
    x = pa.input_from_history(in_sig,filt_size)
    d = desired_sig.reshape(len(desired_sig), 1)
    d = d[:N-filt_size+1, 0]

    f = pa.filters.FilterRLS(n=filt_size, mu=m, w="zeros")
    y, e, w = f.run(d, x)
    return y, e, w  

def filter_wandering_noise(sig):
    d = np.ones(len(sig))
    filt_size = 1
    y, e, _= filter_with_RLS(d, sig, n = filt_size, m = 0.99)
    return e
```
