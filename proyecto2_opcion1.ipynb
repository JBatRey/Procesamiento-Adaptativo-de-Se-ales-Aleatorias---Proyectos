#Implementación de los filtros sugeridos en el paper (implementación I)

```python
def filter_with_RLS(in_sig, desired_sig, n = 1, m = 0.01):
    # creation of data
    N = len(in_sig)
    filt_size =  n
    x = pa.input_from_history(in_sig,filt_size)
    d = desired_sig.reshape(len(desired_sig), 1)
    d = d[:N-filt_size+1, 0]

    f = pa.filters.FilterRLS(n=filt_size, mu=m, w="zeros")
    y, e, w = f.run(d, x)
    
    return y, e , w   

```

```python
#Filtramos el wandering noise
def filter_wandering_noise(sig):
    d = np.ones(len(sig))
    filt_size = 1
    y, e, _= filter_with_RLS(d, sig, n = filt_size, m = 0.99)
    return e

def create_xhz_ref(length, freq, sr):

    N = 2*np.cos(2*np.pi*freq/sr)
    e = np.ones(length)
    e[0]=0.5
    for n in range(2, e.size):
        e[n] = N*e[n-1]-e[n-2]

    return e

def filter_60hz(signal, fs=360):
    ref = create_xhz_ref(len(signal), 60, 360) 
    y, e, w = filter_with_RLS(ref, signal, n = 6, m = 0.97)
    ref = create_xhz_ref(len(e), 120, 360) 
    y, e, w = filter_with_RLS(ref, signal, n = 2, m = 0.99)
    return e

from ecgdetectors import Detectors

def detect_r_peak(sig):
    detectors = Detectors(360)
    r_peaks = np.array(detectors.engzee_detector(sig))
    deltas = np.zeros(len(sig))
    deltas[r_peaks] = 1

    return deltas


def detect_r_peak2(sig):
    detectors = Detectors(360)
    r_peaks = np.array(detectors.hamilton_detector(sig))
    deltas = np.zeros(len(sig))
    deltas[r_peaks] = 1

    return deltas

def arf_filter(signal, filt_size, mu = 0.99, offset = 150):
    desired3 = signal
    deltas_og = detect_r_peak(desired3)
    deltas = np.concatenate((np.zeros(offset),deltas_og))[:-offset]

    y, e, w = filter_with_RLS(deltas, desired3, n = filt_size, m=mu)

    return e, y

def arf_filter2(signal, filt_size, mu = 0.99, offset = 150):
    desired3 = signal
    deltas_og = detect_r_peak2(desired3)
    deltas = np.concatenate((np.zeros(offset),deltas_og))[:-offset]

    y, e, w = filter_with_RLS(deltas, desired3, n = filt_size, m=mu)

    return e, y
```
